```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
data=pd.read_csv('/content/global_freelancers_raw.csv')
```

```python
cat_cols = data.select_dtypes(include='object').columns
data_onehot = pd.get_dummies(data, columns=cat_cols, drop_first=True)
```

```python
data_label = data.copy()
le = LabelEncoder()

for col in cat_cols:
    data_label[col] = le.fit_transform(data_label[col])
```

```python
data_ordinal = data.copy()
ord_enc = OrdinalEncoder()
data_ordinal[cat_cols] = ord_enc.fit_transform(data_ordinal[cat_cols])
```

```python
data_freq = data.copy()

for col in cat_cols:
    freq_map = data_freq[col].value_counts()
    data_freq[col] = data_freq[col].map(freq_map)
```
